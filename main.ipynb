Выгрузка данных в xlsx в папку data

```python
import sqlite3
import pandas as pd

with sqlite3.connect('chinook.db') as conn:
    df_albums = pd.read_sql('select * from albums', conn)
    df_artists = pd.read_sql('select * from artists', conn)
    df_customers = pd.read_sql('select * from customers', conn)
    df_employees = pd.read_sql('select * from employees', conn)
    df_genres = pd.read_sql('select * from genres', conn)
    df_invoice_items = pd.read_sql('select * from invoice_items', conn)
    df_invoices = pd.read_sql('select * from invoices', conn)
    df_media_types = pd.read_sql('select * from media_types', conn)
    df_playlist_track = pd.read_sql('select * from playlist_track', conn)
    df_playlists = pd.read_sql('select * from playlists', conn)
    df_sqlite_sequence = pd.read_sql('select * from sqlite_sequence', conn)
    df_sqlite_stat1 = pd.read_sql('select * from sqlite_stat1', conn)
    df_tracks = pd.read_sql('select * from tracks', conn)

df_albums.to_excel('data/albums.xlsx', index=False)
df_artists.to_excel('data/artists.xlsx', index=False)
df_customers.to_excel('data/customers.xlsx', index=False)
df_employees.to_excel('data/employees.xlsx', index=False)
df_genres.to_excel('data/genres.xlsx', index=False)
df_invoice_items.to_excel('data/invoice_items.xlsx', index=False)
df_invoices.to_excel('data/invoices.xlsx', index=False)
df_media_types.to_excel('data/media_types.xlsx', index=False)
df_playlist_track.to_excel('data/playlist_track.xlsx', index=False)
df_playlists.to_excel('data/playlists.xlsx', index=False)
df_sqlite_sequence.to_excel('data/sqlite_sequence.xlsx', index=False)
df_sqlite_stat1.to_excel('data/sqlite_stat1.xlsx', index=False)
df_tracks.to_excel('data/tracks.xlsx', index=False)
```

Определение средней длительности треков в каждом жанре

```python
import pandas as pd
from readfile import read_file

tracks = read_file('data/tracks.xlsx')
avg_track_duration = tracks.groupby('GenreId')['Milliseconds'].mean().astype(int)
df_avg_track_duration = avg_track_duration.reset_index()

genres = read_file('data/genres.xlsx')
avg_duration = pd.merge(genres, df_avg_track_duration, on='GenreId')
avg_duration.rename(columns={'Milliseconds': 'AverageTrackDuration(Milliseconds)'}, inplace=True)
avg_duration
```

Формирование списка треков с названием альбома и именем исполнителя

```python
tracks = read_file('data/tracks.xlsx')
albums = read_file('data/albums.xlsx')
artists = read_file('data/artists.xlsx')

tracks_albums = pd.merge(tracks, albums, on='AlbumId')
tracks_albums_artists = pd.merge(tracks_albums, artists, on='ArtistId')
tracks_albums_artists.rename(columns={'Name_x': 'Track_name', 'Title': 'Album_title', 'Name_y': 'Artist_name'}, inplace=True)
tracks_albums_artists
tracks_list = tracks_albums_artists[['Track_name', 'Album_title', 'Artist_name']]
tracks_list
```
